```python
import torch

# Check if CUDA is available
print("CUDA available:", torch.cuda.is_available())

# Get the CUDA version
print("CUDA version:", torch.version.cuda)

# Check cuDNN version
print("cuDNN version:", torch.backends.cudnn.version())

# Get GPU details
if torch.cuda.is_available():
    print("Device Name:", torch.cuda.get_device_name(0))
    print("Number of GPUs available:", torch.cuda.device_count())
    print("Current Device ID:", torch.cuda.current_device())
```

```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
# Generate sample data
np.random.seed(42)
x = np.random.rand(100) * 10  # Random values for x-axis
y = np.random.rand(100) * 10  # Random values for y-axis
z = np.random.rand(100) * 10  # Random values for z-axis
colors = np.random.rand(100)  # Color values
```

```python
# Create a 3D plot
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
```

```python

# Scatter plot in 3D
scatter = ax.scatter(x, y, z, c=colors, cmap='viridis', s=50, alpha=0.8)
```

```python
# Add labels
ax.set_title("3D Scatter Plot", fontsize=14)
ax.set_xlabel("X Axis")
ax.set_ylabel("Y Axis")
ax.set_zlabel("Z Axis")
```

```python
# Add a color bar
color_bar = plt.colorbar(scatter, ax=ax, shrink=0.6, aspect=10)
color_bar.set_label('Color Scale')

# Show the plot
plt.show()
```
